```python
from sklearn.model_selection import train_test_split
X=df.drop("benign_0__mal_1",axis=1).values
y=df["benign_0__mal_1"].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler

scale=MinMaxScaler()

X_train=scale.fit_transform(X_train)

X_test=scale.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
layer=list()
for i in  np.linspace(5,30,5):
    layer.append(int(i))
layer
```

```python
layer=np.array(layer)
layer
```

```python
base_mod=Sequential()
```

```python
for i in layer:
    base_mod.add(Dense(i,activation='relu')) # cause relu is the most widely used
# BINARY CLASSIFICATION
base_mod.add(Dense(1,activation='sigmoid'))  # last one is binary classification therefore we  have one output and activation 
#selected is sigmoid here cause we want our output to range between 0,1
```

```python
from sklearn.metrics import accuracy_score,precision_score,classification_report,confusion_matrix
```

```python
base_mod.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
base_mod.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=500) # yaha pe sirf train data daala and batch size nhi dala caus edata is kaafi kam so nhi dala
```

```python
base_mod.layers
```

```python
loss=pd.DataFrame(base_mod.history.history)
loss
```

```python
plt.plot(np.arange(0,500),loss['loss'])
plt.plot(np.arange(0,500),loss['val_loss']) # here we can see that we are oevrfitting as val_loss is increasing eventhough our
#train scores are decreasing.so we can say that teh no of eppoch is very large
```

```python
X_train.shape
```

```python
pred=base_mod.predict(X_test)
pred
y_test
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early=EarlyStopping(monitor='val_loss',mode='min',patience=20) #mode  matlab jo bhi quantity monitor kar rhe ham voh quantity 
#ko maximize karne ka goal hai ki minimize..if kuch nhi dala toh model ko dekhke apne aap it will analyze and set if our goal 
#is to max or minimize..patience is after we have reached iteration jisme val _loss starts increasing again so udke baad we 
#wait for 20 more iterations cauuse we set patience =20 if on an average also in the next 20 iterations also our val_loss 
# increases on an average then we stop at that point and come out of epoch so even if we set zyada epoch there is no fear 
#of overfitting
```

```python
layer=list()
for i in  np.linspace(5,30,5):
    layer.append(int(i))
layer

layer=np.array(layer)
layer

new_mod=Sequential()

for i in layer:
    new_mod.add(Dense(i,activation='relu')) 
# BINARY CLASSIFICATION
new_mod.add(Dense(1,activation='sigmoid'))

from sklearn.metrics import accuracy_score,precision_score,classification_report,confusion_matrix

new_mod.compile(loss='binary_crossentropy',optimizer='adam')

new_mod.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=500,callbacks=[early])
```

```python
loss=pd.DataFrame(new_mod.history.history)
loss
```

```python
plt.plot(np.arange(0,48),loss['loss'])
plt.plot(np.arange(0,48),loss['val_loss'])
```

## now if we want to further increase the performaance with keeping the early stop and squeeze a bit more of prformance then what we can do is we can add  a dropout after each layer so that in every epoch the  dropout we mention every epoch some percent of our neurons stay off randomly and aise sare epoch mai randomly uss layer ke utne neuron would stay shut thus training our model for little bit more epoch with earlystopping thus squeezing a bit more performance


## experiment with various dropout percentages ..Typical values is 50 percent if you have two big hidden layers or if you have  multiple small layers then go for small valuees probably..experiment with it more

```python
layer=list()
for i in  np.linspace(5,30,5):
    layer.append(int(i))
layer

layer=np.array(layer)
layer

final_mod=Sequential()

for i in layer:
    final_mod.add(Dense(i,activation='relu'))
    final_mod.add(Dropout(0.1)) 
# BINARY CLASSIFICATION
final_mod.add(Dense(1,activation='sigmoid'))

from sklearn.metrics import accuracy_score,precision_score,classification_report,confusion_matrix

final_mod.compile(loss='binary_crossentropy',optimizer='adam')

final_mod.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=500,callbacks=[early])
```

```python
loss=pd.DataFrame(final_mod.history.history)
loss

plt.plot(np.arange(0,98),loss['loss'])
plt.plot(np.arange(0,98),loss['val_loss'])
```

```python
pred=final_mod.predict_classes(X_test)
pred=pred.reshape(len(pred))
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```
