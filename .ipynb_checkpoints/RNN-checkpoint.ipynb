```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import os
import pandas as pd
```

```python
x=np.linspace(0,100,1000)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python
df=pd.DataFrame(data=y,index=x,columns=['sin_val'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
test_len=0.1
```

```python
len(df)*test_len
```

```python
train=df.iloc[:900]
```

```python
test=df.iloc[900:]
```

```python
train
```

```python
test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
```

```python
scaled_train=scale.fit_transform(train)
```

```python
scaled_test=scale.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length=10
batch_size=1
generator=TimeseriesGenerator(data=scaled_train,targets=scaled_train,length=length,batch_size=batch_size)
```

```python
len(generator)
```

```python
X,y=generator[0]
```

```python
X
```

```python
y
```

```python
n_features=1
from tensorflow.keras.layers import Dense,SimpleRNN,LSTM
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
model.add(SimpleRNN(10,input_shape=(length,n_features)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
model.fit_generator(generator,epochs=10)
```

```python
history=model.history.history
```

```python
plt.plot(history['loss'])
```

```python
eval_batch=scaled_train[-length:]
```

```python
eval_batch
```

```python
reshaped=eval_batch.reshape((1,length,n_features))
```

```python
model.predict(reshaped)
```

```python
test_pred=[]
eval_batch=scaled_train[-length:]

reshaped=eval_batch.reshape((1,length,n_features))


for i in range(len(scaled_test)):
    current=model.predict(reshaped)[0]
    test_pred.append(current)
    reshaped=np.append(reshaped[:,1:,:],[[current]],axis=1)
    
```

```python
test_pred
```

```python
og_test_pred=scale.inverse_transform(test_pred)
```

```python
test.head()
```

```python
og_test_pred[:5]
```

```python
test['pred']=og_test_pred
```

```python
test
```

```python
plt.plot(test['sin_val'],'b')
plt.plot(test['pred'],'r')
```
