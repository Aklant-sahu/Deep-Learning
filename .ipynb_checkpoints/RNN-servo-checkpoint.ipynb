```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import os
import pandas as pd
```

```python
x=np.arange(0,1000)
```

```python
x
```

```python
y=[]
val=0
inc=1
dec=0
for i in range(len(x)):
    if(inc==1 ):
        val+=5
        y.append(val)
        if val==180:
            inc=0
            dec=1
    elif(dec==1):
        val-=5
        y.append(val)
        if val==0:
            inc=1
            dec=0
            
        
    
        
```

```python
y
```

```python
plt.plot(x,y)
```

```python
df=pd.DataFrame(data=y,columns=['pos_val'],index=x)
```

```python
df
```

```python
test_len=0.1

len(df)*test_len
```

```python
train=df.iloc[:900]
```

```python
train
```

```python




test=df.iloc[900:]

test
```

```python
from sklearn.preprocessing import MinMaxScaler

scale=MinMaxScaler()

scaled_train=scale.fit_transform(train)

scaled_test=scale.transform(test)

from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator

length=10
batch_size=1
generator=TimeseriesGenerator(data=scaled_train,targets=scaled_train,length=length,batch_size=batch_size)

len(generator)
```

```python
X,y=generator[0]

X

y

n_features=1
from tensorflow.keras.layers import Dense,SimpleRNN,LSTM
from tensorflow.keras.models import Sequential

model=Sequential()
model.add(SimpleRNN(10,input_shape=(length,n_features)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')

model.summary()
```

```python
model.fit_generator(generator,epochs=5)
```

```python
eval_batch=scaled_train[-length:]

eval_batch

reshaped=eval_batch.reshape((1,length,n_features))

model.predict(reshaped)
```

```python
test_pred=[]
eval_batch=scaled_train[-length:]

reshaped=eval_batch.reshape((1,length,n_features))


for i in range(len(scaled_test)):
    current=model.predict(reshaped)[0]
    test_pred.append(current)
    reshaped=np.append(reshaped[:,1:,:],[[current]],axis=1)
    
```

```python
test_pred
```

```python
og_test_pred=scale.inverse_transform(test_pred)
```

```python
test.head(15)
```

```python
og_test_pred[:15]
```

```python
test['pred']=og_test_pred
```

```python
plt.plot(test['pos_val'],'b')
plt.plot(test['pred'],'r')
```

```python
X,y=generator[0]

X

y

n_features=1
from tensorflow.keras.layers import Dense,SimpleRNN,LSTM
from tensorflow.keras.models import Sequential

lstm_model=Sequential()
lstm_model.add(LSTM(10,input_shape=(length,n_features)))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam',loss='mse')

lstm_model.summary()
```

```python
lstm_model.fit_generator(generator,epochs=5)
```

```python


test_pred=[]
eval_batch=scaled_train[-length:]

reshaped=eval_batch.reshape((1,length,n_features))


for i in range(len(scaled_test)):
    current=lstm_model.predict(reshaped)[0]
    test_pred.append(current)
    reshaped=np.append(reshaped[:,1:,:],[[current]],axis=1)
    



og_test_pred=scale.inverse_transform(test_pred)


test['pred']=og_test_pred
```

```python
plt.plot(test['pos_val'],'b')
plt.plot(test['pred'],'r')
```
